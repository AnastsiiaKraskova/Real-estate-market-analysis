## Предобработка данных

Начнем обработку пропусков с простого. Пропуски в колонках `is_apartment`, `open_plan`  вероятнее всего означают, что объект не является апартаментом, не имеет свободной планировки.  Заменим пропуски в этих колнках на *False*.

Пропуски в колонке `balcony`,  `parks_around3000`, `ponds_around3000` вероятнее всего означают, что балкона нет, нет парков в радиусе 3 км и нет водоемов  в радиусе 3 км. Заменим такие пропуски на 0.

Вероятно при заполнении указанных колонок их просто пропустили, чтобы не стаить 0 или "нет", поэтому наша замена логична.

```python
data['is_apartment'] = data['is_apartment'].fillna('False') # меняем пропуски в колонке is_apartment на False
data['open_plan'] = data['open_plan'].fillna('False') # меняем пропуски в колонке open_plan на False
data['balcony'] = data['balcony'].fillna(0) # меняем пропуски в колонке balcony на 0
data['parks_around3000'] = data['parks_around3000'].fillna(0) # меняем пропуски в колонке parks_around3000 на 0
data['ponds_around3000'] = data['ponds_around3000'].fillna(0)# меняем пропуски в колонке ponds_around3000 на 0
```

### Обработка значений в колонке `ceiling_height`

Перейдем к обработке значений в колонке `ceiling_height`. По общей статистике мы увидели, что в этой колонке есть неадекватные минимальные и максимальные значения. Посмотрим на минимальные значения.

```python
data.sort_values(by='ceiling_height', ascending=True).head(15)
```

По первым 15 строчкам осортированным по возрастанию, видим, что высота потолков от 1 метра до 2, только в 14 объектах. Допустим, что потолки ниже 2.2 быть не могут. Посмотрим на максимальные значения.

```python
data.sort_values(by='ceiling_height', ascending=False).head(35)
```

Достаточно вывести 35 строк отсортированных по высоте потолка по убыванию, чтобы понять, что высота потолка 100 это ошибка, но это единичиный случай. 21 строка содержат значения от 32 до 22,6, похоже на ошибку пунктуации (точка не там), исправим перенеся точку на 1 цифру назад, чтобы из 22,6 получилось 2,26.

```python
import numpy as np #библиотека для замены некорректных значений на NaN

#заменим значения, где точка не там.
for i in range(len(data['ceiling_height'])): 
    if data['ceiling_height'][i] > 20.0:
        data['ceiling_height'][i]=data['ceiling_height'][i]/10


mistake = [10.0, 20.0, 14.0, 10.3, 8.3, 8.0, 1.0, 1.2, 1.75, 2.0]
data['ceiling_height'] = data['ceiling_height'].replace(mistake, np.nan)
data.sort_values(by='ceiling_height', ascending=False).head()
```

Высота потолков в квартирах регламентируется нормативными документами СНиП, согласно которым на высоту влияет регион строительства и площадь из тех данных, что у нас есть. Учтем еще этажность, свободная планировка и апртаменты ли это. На основании 5 параметров заменим все NaN в колонке `ceiling_height`. Для начала создадим столбец с категорией по  общей площади `group_total_area`. 

```python
def total_area_group(area): #создаем функцию total_area_group, для создании категорий
    if area <= 40:
        return 'меньше 40'
    if area <= 55:
        return 'от 41 до 55'
    if area <= 70:
        return 'от 56 до 70'
    if area <= 100:
        return 'от 71 до 100'
    if area <= 200:
        return 'от 101 до 200'
    if area <= 300:
        return 'от 201 до 300'
    if area <= 400:
        return 'от 301 до 400'
    if area <= 500:
        return 'от 401 до 500'
    if area <= 600:
        return 'от 501 до 600'
    if area <= 700:
        return 'от 601 до 700'
    if area <= 800:
        return 'от 701 до 800'
    return 'больше 800'

data['group_total_area'] = data['total_area'].apply(total_area_group)#добавляем новый столбец 
# с категорией площади

data.head()#проверяем, что столбец добавился верно 
```

Заменим пропуски в значениях высота потолка на основании 6 параметров и выведем сколько после замены осталось пустых значений. Будем уменьшать количество параметров, пока не заменим все пропуски.

```python
data['ceiling_height'] = data['ceiling_height'].fillna(data.groupby(['group_total_area', 'floors_total',
                                                                     'is_apartment', 'locality_name', 'open_plan', 'studio'])
                                                                    ['ceiling_height'].transform('median'))
display ('Количество пропусков в колонке "высота потолка" после замены по 6 параметрам', data['ceiling_height'].isna().sum())
data['ceiling_height'] = data['ceiling_height'].fillna(data.groupby(['group_total_area', 'floors_total', 
                                                                     'is_apartment', 'open_plan'])['ceiling_height']
                                                                    .transform('median'))
display ('Количество пропусков в колонке "высота потолка" после замены по 4 параметрам', data['ceiling_height'].isna().sum())
data['ceiling_height'] = data['ceiling_height'].fillna(data.groupby(['group_total_area', 'floors_total'])
                                                                    ['ceiling_height'].transform('median'))
display ('Количество пропусков в колонке "высота потолка" после замены по 2 параметрам', data['ceiling_height'].isna().sum())
data['ceiling_height'] = data['ceiling_height'].fillna(data.groupby(['group_total_area'])['ceiling_height'].transform('median'))
display ('Количество пропусков в колонке "высота потолка" после замены по 1 параметру', data['ceiling_height'].isna().sum())


data.sort_values(by='ceiling_height')# проверим что диапазон высот после замен адекватный
```

### Обработка значений в колонках `living_area`

По минимальным значения колонки `living_area`ясно, что есть достаточное количество искаженных данных. Посмотрим на отношение жилой площади к общей и как эти значения распределяются на графике

```python
data['living_area_ratio'] = data['living_area']/data['total_area'] #добавим колонку с соотношением площади жилой к общей
data.hist('living_area_ratio', bins=20) #построим гистограмму отношения
```

По графику видно, что распределение нормальное, но с логической точки зрения отношение жилой к общей не может быть слишком маленьким и слишком большим. По графику видно, что отношение меньше 0,4 уже редкость, поэтому исправим эти данные на медианные с условием группировки, аналогичной высоте потолка. Тоже проделаем со значениями больше 0,7.

```python
# заменим значения жилой площади, где отношение жилой к общей, меньше 0.4 на медианнные значения с группировкой по 6 параметрам

display (data['living_area'].isna().sum())
data.loc[(data.living_area_ratio <= 0.4), 'living_area'] = np.nan
data.loc[(data.living_area_ratio >= 0.7), 'living_area'] = np.nan

display (data['living_area'].isna().sum())
data['living_area_ratio'] = data['living_area']/data['total_area'] #обновим данные в колонке после замен
data.sort_values(by='living_area_ratio', ascending=False).head()
```

```python
#заменим пропуски в жилой площади на мединное значения отношения к общей площади*общую площадь        
data['living_area'] = data['living_area'].fillna((data['living_area_ratio'].median())*data['total_area']).round(2)   
data['living_area_ratio'] = data['living_area']/data['total_area'] #обновим колонку после замены
data.sort_values(by='living_area', ascending=False).head(15)
```

С пропусками остались колонки `airports_nearest` `cityCenters_nearest` `parks_nearest` `ponds_nearest` `days_exposition` `floors_total` `locality_name` заменить эти пропуски не удасться, т.к. нет подходящего значения на замену. К примеру, для расстояния до аэропорта нужен адрес, не верно будет решить что расстояние до аэропорта такое же как в других квартира в том же городе и т.д.

Заменим типы данных в колонках и перейдем к анализу.

В колонке `first_day_exposition` поменяем тип на дату и время, чтобы потом извлечь день, месяц и год для исследования.
В колонке `is_apartment` заменим тип на булев, чтобы использовать в анализе.
           
В колонках ниже, заменим тип на целочисленный, чтобы отбросить лишние нули после запятой:
* `last_price`
* `balcony`
* `days_exposition` (чтобы не было ошибки из-за пропусков, сначала пропуски поменям на 0, т.к. объектов проданных за 0 дней нет)

```python
data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'], format='%Y.%m.%dT%H:%M:%S')
data['is_apartment'] = data['is_apartment'].astype('bool')

data['days_exposition'].fillna(0, inplace=True)

change = ['last_price', 'balcony', 'days_exposition']
for column in change:
    data[column] = data[column].astype('int')
```

Итак, высоту потолков привели в адекватный вид, теперь диапазон потолков от 2,2 до 6 метров. Пропуски в жилой площади заменили  на мединное значение отношения к общей площади умножить на общую площадь. Заранее посчитав, что отношение жилой к общец должно находиться в диапазоне от 0,4 до 0,7. Пропуски в площади кухни исправили аналогично жилой, только диапазон отношения площади кухни к общей от 0,1 до 0,3. Заменили тип данных в колонке `first_day_exposition` на datetime. В колонках `last_price`, `balcony`, `days_exposition` заменили тип на целочисленный, чтобы отбросить лишние нули после запятой.

### Определение центральной зоны в Санкт-Петербурге

Выделим квартиры в Санкт-Петербурге (`locality_name`). Задача — выяснить, какая область входит в центр. Для этого создадим столбец с расстоянием до центра в километрах: округлим до целых значений. После этого посчитаем среднюю цену для каждого километра. Построим график: он должен показывать, как цена зависит от удалённости от центра. Определим границу, где график сильно меняется, — это и будет центральная зона.

```python
data_sp = data.query('locality_name == "Санкт-Петербург"') #создаем таблицу с данными только по Санкт-Петербургу
data_sp['cityCenters_nearest_km'] = (data_sp['cityCenters_nearest']/1000).round(0)
#добавляем колонку с удаленностью от центра в км
data_sp.info()
```

```python
km = data_sp.groupby("cityCenters_nearest_km")["price_per_metr2"].mean().round(0)
#создаем таблицу со средней ценой за каждый км от центра
```

```python
km.plot(kind="barh", fontsize=10, figsize=(10, 15))
plt.show()
```

Средняя цена за киллометр резко падает на значении 3 км, второй провал примерно на том же уровне что и у 3х км находится на значении 8 км. Т.к. Санкт-Петербург крупный город будем считать центральной зоной квартиры с удаленнностью не более 8 км.
